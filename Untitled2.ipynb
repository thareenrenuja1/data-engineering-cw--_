```python
import boto3
import pandas as pd
from io import StringIO
import time

AWS_ACCESS_KEY ="AKIA3RIUTS6JUF7PJBP4"
AWS_SECRET_KEY = "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"
AWS_REGION = "ap-south-1"
SCHEMA_NAME = "covid_19_iit"
S3_STAGING_DIR ="s3://darshil-covid-de-project-iit-output/Unsaved/"
S3_BUCKET_NAME ="darshil-covid-de-project-iit-output"
S3_OUTPUT_DIRECTORY ="Unsaved"

athena_client = boto3.client(
    "athena",
    aws_access_key_id=AWS_ACCESS_KEY,
    aws_secret_access_key=AWS_SECRET_KEY,
    region_name=AWS_REGION,
)

Dict ={}

def download_and_load_query_results(client: boto3.client, query_response: Dict) -> pd.DataFrame:
    while True:
        try:
            # This function only loads first 1000 rows
            query_execution = client.get_query_execution(QueryExecutionId=query_response["QueryExecutionId"])
            status = query_execution['QueryExecution']['Status']['State']
            if status in ['SUCCEEDED', 'FAILED', 'CANCELLED']:
                break
            time.sleep(0.001)
        except Exception as err:
            if "not yet finished" in str(err):
                time.sleep(0.001)
            else:
                raise err

    temp_file_location: str ="athena_query_results.csv"
    s3_client = boto3.client(
        "s3",
        aws_access_key_id=AWS_ACCESS_KEY,
        aws_secret_access_key=AWS_SECRET_KEY,
        region_name=AWS_REGION,
    )
    
    s3_client.download_file(
           S3_BUCKET_NAME,
           f"{S3_OUTPUT_DIRECTORY}/{query_response['QueryExecutionId']}.csv",
           temp_file_location,
    )
    
    return pd.read_csv(temp_file_location)

response = athena_client.start_query_execution(
    QueryString="SELECT * FROM enigma_jhud",
    QueryExecutionContext={"Database": SCHEMA_NAME},
    ResultConfiguration={
        "OutputLocation": S3_STAGING_DIR,
        "EncryptionConfiguration": {"EncryptionOption": "SSE_S3"},
    },
)

enigma_jhud = download_and_load_query_results(athena_client, response)
enigma_jhud.head()
```

```python
dimDate = rearc_covid_19_testing_data_states_dailystates_daily[['fips','date']] 
dimDate.head()
```

```python
import pandas as pd

# Assuming 'dimDate' is the DataFrame containing the 'fips' and 'date' columns

# Convert 'date' column to string format
dimDate['date'] = dimDate['date'].astype(str)

# Extract year, month, and day components from the string format
dimDate['year'] = dimDate['date'].str[:4]
dimDate['month'] = dimDate['date'].str[4:6]
dimDate['day'] = dimDate['date'].str[6:]

# Concatenate year, month, and day components to form a date string
dimDate['date'] = dimDate['year'] + '-' + dimDate['month'] + '-' + dimDate['day']

# Convert 'date' column to datetime format
dimDate['date'] = pd.to_datetime(dimDate['date'])

# Drop the temporary year, month, and day columns
dimDate.drop(['year', 'month', 'day'], axis=1, inplace=True)

# Display the updated 'dimDate' DataFrame
dimDate.head()
```

```python
dimDate['year'] = dimDate['date'].dt.year
dimDate['month'] = dimDate['date'].dt.month
dimDate['day_of_week'] = dimDate['date'].dt.dayofweek
```
